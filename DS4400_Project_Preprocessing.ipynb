```python
import sklearn
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('smoking_driking_dataset_Ver01.csv')
print(df.head(20))
```

```python
# Replacing: "Y" ->  1, "N" -> 0
df['DRK_YN'] = df['DRK_YN'].replace("Y", 1).replace("N",0)
# Replacing: "Male" ->  0, "Female" -> 1
df['sex'] = df['sex'].replace("Male", 0).replace("Female", 1)
# Replacing: abnormal 2 -> abnormal 0
df['hear_left'] = df['hear_left'].replace(2, 0)
# Replacing: abnormal 2 -> abnormal 0
df['hear_right'] = df['hear_right'].replace(2, 0)
# Replacing: never smoked 1 -> non-smoker 0, used to smoke -> non-smoker 0, still smokes -> smoker 1
df['SMK_stat_type_cd'] = df['SMK_stat_type_cd'].replace(1, 0).replace(2, 0).replace(3, 1)

# Uncomment to make age 50 and under
# df = df.loc[df['age'] <= 50]
# Uncomment to make age above 60
# df = df.loc[df['age'] > 50]

print(df.head(20))
```

```python
# Ignoring the target values as well as categorical variables
df_features = df.drop(columns=['sex', 'urine_protein', 'hear_left', 'hear_right', 'DRK_YN', 'SMK_stat_type_cd'])

# Scaling the data
scaler = StandardScaler()
X = scaler.fit_transform(df_features)
scaled_df = pd.DataFrame(X, columns=df_features.columns)

# Adding back the values
scaled_df['sex'] = df['sex'].values
scaled_df['urine_protein'] = df['urine_protein'].values
scaled_df['hear_left'] = df['hear_left'].values
scaled_df['hear_right'] = df['hear_right'].values
scaled_df['DRK_YN'] = df['DRK_YN'].values
scaled_df['SMK_stat_type_cd'] = df['SMK_stat_type_cd'].values

print(scaled_df.head(20))

scaled_df.to_csv('scaled_smoking_drinking_dataset.csv', index=False)
# scaled_df.to_csv('scaled_smoking_drinking_dataset_50_and_under.csv', index=False)
# scaled_df.to_csv('scaled_smoking_drinking_dataset_above_50.csv', index=False)
```
